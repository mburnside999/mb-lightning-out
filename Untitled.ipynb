```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Sample data for training
X = np.array([[1, 2], [2, 3], [3, 1], [6, 4], [5, 6], [7, 7]])
y = np.array([0, 0, 0, 1, 1, 1])

# Create an SVM classifier with a linear kernel
classifier = svm.SVC(kernel='linear')

# Train the classifier
classifier.fit(X, y)

# Get the separating hyperplane
w = classifier.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(min(X[:, 0]), max(X[:, 0]), 10)
yy = a * xx - (classifier.intercept_[0]) / w[1]

# Plot the points and the hyperplane
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.plot(xx, yy, 'k-')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Linear Classification')
x = 2
y = 3
# Create a scatter plot with the single point
plt.scatter(x, y, c='blue', label='Single Point')

x = 6
y = 2
# Create a scatter plot with the single point
plt.scatter(x, y, c='green', label='Single Point')


plt.show()
print ('prediction for 2,3 = ',classifier.predict([[2,3]]))
print ('prediction for 6,2 = ',classifier.predict([[6,2]]))
```
