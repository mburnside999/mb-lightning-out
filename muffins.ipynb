```python
#data analysis
import pandas as pd
import numpy as np
#support vector
from sklearn import svm

#visualisation
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2)
```

```python
recipes=pd.read_csv('muffins.csv')
```

```python
print(recipes.head())
```

```python
#plot sugar and flour
sns.lmplot(data=recipes,x='Flour',y='Sugar',hue='Type',palette='Set1',fit_reg=False);
x = [40]
y = [10]
plt.plot(x, y, marker="o", markersize=10, markeredgecolor="blue", markerfacecolor="blue")

plt.show()

```

```python
type_label=np.where(recipes['Type']=='Muffin',0,1)
```

```python
print (type_label)
```

```python
recipe_features=recipes.columns.values[1:].tolist()
recipe_features
```

```python
ingredients=recipes[['Flour','Sugar']].values
```

```python
print (ingredients)
```

```python
model=svm.SVC(kernel='linear',C=1.0)
model.fit(ingredients,type_label)
```

```python
# get the separting hyperplane
w=model.coef_[0]
print(w)
```

```python
print (w)
```

```python
print (-w[1])
```

```python
b=model.intercept_[0]
print (b)
```

```python
x_points = np.linspace(35, 55) 
print (x_points)
y_points = -(w[0] / w[1]) * x_points - b / w[1] 
print (y_points)
```

```python
sns.lmplot(data=recipes,x='Flour',y='Sugar',hue='Type',palette='Set1',fit_reg=False);
```

```python
w = model.coef_[0]
print(w)

a = -w[0] / w[1]

xx = np.linspace(44,55)
yy = a * xx - model.intercept_[0] / w[1]

h0 = plt.plot(xx, yy, 'k-', label="non weighted div")
```
